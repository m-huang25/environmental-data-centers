```r
# convert .csv to a dataframe
df <- read.csv("data/raw_data/issued_air_permits.csv", header = FALSE)

# assign correct column names
colnames(df) <- c(
  "air_site_name",
  "registration_no.",
  "permit_issuance_date",
  "program_type",
  "city_or_county",
  "regional_office"
)

```

```r
# display the dataframe
head(df,10)
```

```r
# checking the column names
names(df)
```

```r
# Check the exact values in city_or_county for extra spaces or different capitalizations
unique(df$city_or_county)
```

```r
library(stringr)

# trim spaces from city_or_county
df <- df %>%
  mutate(city_or_county = str_trim(city_or_county))  # Trim spaces

# Verify the cleaned column values
unique(df$city_or_county)
```

```r
#filter to include only Loudoun County, Prince William County, and Fairfax County; use of the operator %in% to check if elements of one vector are present in another vector. It returns a logical vector (TRUE or FALSE) for each element, and only rows where it’s TRUE are returned.
local_air_permits <- df %>%
  filter(city_or_county %in% c("Loudoun County", "Prince William County", "Fairfax County"))
local_air_permits
```
