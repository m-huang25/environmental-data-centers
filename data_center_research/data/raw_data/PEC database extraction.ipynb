```python
## First attempt to retrieve data from PEC website

import requests
import pandas as pd

url = "https://services3.arcgis.com/mTaShYKffyWc5uRb/arcgis/rest/services/Data_Centers_Virginia/FeatureServer?f=json"

params = {
    "where": "1=1",          # get all rows
    "outFields": "*",        # all columns
    "f": "json",
    "returnGeometry": "false"
}

response = requests.get(url, params=params)
data = response.json()

# Extract attributes and convert to DataFrame
rows = [feature["attributes"] for feature in data["features"]]
df = pd.DataFrame(rows)

# Save CSV
df.to_csv("arcgis_table.csv", index=False)

print("CSV saved! Number of rows:", len(df))
df.head()
```

```python
## Third attempt to retrieve data from PEC website - successful

import requests
import pandas as pd

# Updated URL with layer 19
url = "https://services3.arcgis.com/mTaShYKffyWc5uRb/arcgis/rest/services/Data_Centers_Virginia/FeatureServer/19/query"

params = {
    "where": "1=1",          # get all rows
    "outFields": "*",        # all columns
    "f": "json",             # specify JSON response
    "returnGeometry": "false"  # no geometry, just the data
}

response = requests.get(url, params=params)

# Print the raw response to check the content
print("Response Status Code:", response.status_code)
print("Response Text:", response.text[:500])  # Print the first 500 characters

# Try parsing JSON
try:
    data = response.json()
    
    # Check if there are any features in the data
    if "features" in data:
        rows = [feature["attributes"] for feature in data["features"]]
        df = pd.DataFrame(rows)
        df.to_csv("arcgis_table.csv", index=False)
        print("CSV saved!")
        print("Data Sample:\n", df.head())
    else:
        print("No features found in the response. Data may be empty or incorrectly formatted.")
except Exception as e:
    print("Error decoding JSON:", e)

```
